## 2. Body Mass Index (BMI) Calculation

### 2.1 Calculate Female BMI

This step involves computing the Body Mass Index (BMI) for each female participant. BMI is calculated as weight in kilograms divided by the square of height in meters. The resulting BMI values are then appended as a new column to the `female` matrix.

```python
# Extract weight (kg) which is the first column (index 0)
female_weight_kg = female[:, 0]

# Extract standing height (cm) which is the second column (index 1)
female_height_cm = female[:, 1]

# Convert height from centimeters to meters
female_height_m = female_height_cm / 100.0

# Calculate BMI: weight (kg) / (height (m))^2
female_bmi = female_weight_kg / (female_height_m ** 2)

# Reshape the BMI array to be a column vector for concatenation
female_bmi_column = female_bmi.reshape(-1, 1)

# Add the BMI column as the eighth column to the female matrix
# np.hstack is used to stack arrays in sequence horizontally (column-wise)
female = np.hstack((female, female_bmi_column))

print(f"Shape of female matrix after adding BMI: {female.shape}")
# Display the first few rows of the updated female matrix to verify
print("First 5 rows of female matrix with BMI:")
print(female[:5, :])
```

#### Summary of Results (2.1)

The Body Mass Index (BMI) has been successfully calculated for female participants and appended as the eighth column to the `female` matrix. The updated shape of the matrix confirms the addition of this new metric, which provides a standardized measure of weight relative to height.

### 2.2 Calculate Male BMI

## 5. Correlation and Standardized Data Analysis

### 5.1 Standardize Female Body Measurements (Z-scores)

To facilitate comparative analysis, each column of the `female` matrix is standardized by computing its z-scores. This transformation results in a new matrix, `zfemale`, where each column has a mean of 0 and a standard deviation of 1, effectively scaling measurements to a common reference point while handling missing values.

```python
import numpy as np

# Create a copy of the female matrix to store standardized values
zfemale = np.copy(female)

# Get the number of columns in the female matrix
num_columns = female.shape[1]

# Iterate through each column to standardize it
for col_idx in range(num_columns):
    # Extract the current column
    column_data = female[:, col_idx]

    # Calculate mean and standard deviation, ignoring NaN values
    # np.nanmean and np.nanstd handle NaN values by excluding them from the calculation
    mean_val = np.nanmean(column_data)
    std_val = np.nanstd(column_data)

    # Standardize the column (compute z-score)
    # (value - mean) / standard deviation
    # Division by zero for columns with std_val = 0 (e.g., all same value or all NaNs) will result in NaN or inf
    # np.seterr(divide='ignore', invalid='ignore') can suppress warnings, but it's often better to handle explicitly.
    # Here, if std_val is 0, the z-score will be 0 or NaN, which is appropriate.
    if std_val == 0:
        # If standard deviation is 0, all non-NaN values are the same as the mean
        # So, their z-score is 0. NaN values remain NaN.
        standardized_column = np.full_like(column_data, np.nan, dtype=float)
        standardized_column[~np.isnan(column_data)] = 0.0
    else:
        standardized_column = (column_data - mean_val) / std_val

    # Update the corresponding column in zfemale
    zfemale[:, col_idx] = standardized_column

print("Shape of zfemale matrix:", zfemale.shape)
print("First 5 rows of zfemale matrix (standardized female data):")
print(zfemale[:5, :])
```

#### Summary of Results (5.1)

The `zfemale` matrix has been successfully created, containing the standardized (z-scored) body measurements for female participants. This transformation is crucial for analyses that require features to be on a comparable scale, effectively converting raw measurements into standard deviation units from their respective means.

### 5.2 Analyze Relationships and Correlations in Standardized Female Data

```python
import numpy as np

# Create a copy of the female matrix to store standardized values
zfemale = np.copy(female)

# Get the number of columns in the female matrix
num_columns = female.shape[1]

# Iterate through each column to standardize it
for col_idx in range(num_columns):
    # Extract the current column
    column_data = female[:, col_idx]

    # Calculate mean and standard deviation, ignoring NaN values
    # np.nanmean and np.nanstd handle NaN values by excluding them from the calculation
    mean_val = np.nanmean(column_data)
    std_val = np.nanstd(column_data)

    # Standardize the column (compute z-score)
    # (value - mean) / standard deviation
    # Division by zero for columns with std_val = 0 (e.g., all same value or all NaNs) will result in NaN or inf
    # np.seterr(divide='ignore', invalid='ignore') can suppress warnings, but it's often better to handle explicitly.
    # Here, if std_val is 0, the z-score will be 0 or NaN, which is appropriate.
    if std_val == 0:
        # If standard deviation is 0, all non-NaN values are the same as the mean
        # So, their z-score is 0. NaN values remain NaN.
        standardized_column = np.full_like(column_data, np.nan, dtype=float)
        standardized_column[~np.isnan(column_data)] = 0.0
    else:
        standardized_column = (column_data - mean_val) / std_val

    # Update the corresponding column in zfemale
    zfemale[:, col_idx] = standardized_column

print("Shape of zfemale matrix:", zfemale.shape)
print("First 5 rows of zfemale matrix (standardized female data):")
print(zfemale[:5, :])
```
